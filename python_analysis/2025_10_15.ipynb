```python
# Flatten into DataFrame
rows = []
for params, results in trials:
    row = {}
    row.update(params)
    for k, v in results.items():
        if np.ndim(v) == 0 or (isinstance(v, np.ndarray) and v.size == 1):
            row[k] = float(np.ravel(v)[0])
    rows.append(row)
df = pd.DataFrame(rows)
```
