```julia
using Distributions
```

```julia
# Generate a Poisson distribution with rate 
 = 1.0  # rate parameter
poisson_dist = Poisson()
# Sample from the distribution
samples = rand(poisson_dist, 1000)
# Calculate the mean and variance
mean_val = mean(samples)
var_val = var(samples)
println("Mean: $mean_val, Variance: $var_val")
# plot histogram
using Plots
histogram(samples, title="Poisson Distribution (=$)", xlabel="Value", ylabel="Frequency")
```

```julia
# Generate a Poisson point process (homogeneous) with constant rate  over time interval [0, T] with time step dt
 = 10.0  # rate parameter (events per unit time)
T = 100.0  # total time
dt = 0.1  # time step
time_points = 0:dt:T
# Generate Poisson events
d = Poisson( * dt)
poisson_events = rand.(d, length(time_points)) ./ dt
# Plot the Poisson point process
using Plots
plot(time_points, poisson_events, seriestype=:stem, title="Poisson Point Process (=$)", xlabel="Time", ylabel="Number of Events")
```
